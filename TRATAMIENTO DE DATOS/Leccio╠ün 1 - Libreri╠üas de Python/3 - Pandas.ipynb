## 1 - ¿Qué es pandas?

Pandas es una biblioteca de software de código abierto escrita para el lenguaje de programación Python, que proporciona herramientas y estructuras de datos de alto rendimiento y fáciles de usar para el análisis y la manipulación de datos. Es especialmente popular en la comunidad de ciencia de datos y se utiliza ampliamente para la limpieza, transformación y análisis de datos. Básicamente, es el Excel de Python.


**Características Principales de Pandas**


1. **Estructuras de datos principales**:

    + Series: Una estructura unidimensional similar a una columna en una tabla. Puede contener datos de cualquier tipo.
    + DataFrame: Una estructura bidimensional similar a una tabla en una base de datos o una hoja de cálculo de Excel. Está compuesto por una colección de Series.


2. **Manipulación de datos**:

    + Selección y filtrado de datos.
    + Manipulación y limpieza de datos.
    + Agregación y transformación de datos.
    + Combinar y fusionar conjuntos de datos.


3. **Manejo de datos nulos**:

    + Identificación y manejo de datos nulos.
    + Interpolación y llenado de datos nulos.


4. **Operaciones en datos temporales**:

    + Manipulación de series temporales y fechas.
    + Re-muestreo y agrupación de datos temporales.

## 2 - Estructuras de datos

### 2.1 - Series

Una Serie de pandas es una estructura de datos unidimensional que puede contener datos de cualquier tipo, enteros, flotantes o cadenas. Hay que distinguir entre el tipo de dato de los elementos y el tipo de dato de la propia serie, ésta tendrá el tipo de dato más general. Una serie es similar a una columna en una tabla o una matriz unidimensional en otros lenguajes de programación. Cada elemento en una Serie tiene una etiqueta, conocida como índice, que puede ser utilizado para acceder a los datos de la Serie.


````python
lista = [(3.4 + i)**2 for i in range(10)]


serie = pd.Series(lista)

serie
````

Vemos que todos los elementos de la serie, además de estar indexados, son números reales, float, y el tipo de dato de la propia serie, el dtype, también es float. ¿Qué ocurriría si introducimos una string en la serie?. Veamos que ocurre con el tipo de dato.

````python
serie[2] = 'hola'

serie
````

````python
type(serie[0])
````

El tipo de dato de la serie ha cambiado a `object`, básicamente string, pero sigue conteniendo números reales en su interior. Veamos algunos métodos y atributos de las series.

````python
serie.head()
````

````python
serie.head(7)
````

````python
serie.tail()
````

````python
serie.tail(3)
````

````python
serie.index
````

### 2.2 - DataFrame

Un objeto del tipo DataFrame define un conjunto de datos estructurado en forma de tabla donde cada columna es un objeto de tipo Series, es decir, todos los datos de una misma columna son del mismo tipo, y las filas son registros que pueden contender datos de distintos tipos.

Un DataFrame contiene dos índices, uno para las filas y otro para las columnas, y se puede acceder a sus elementos mediante los nombres de las filas y las columnas. Podemos crear un dataframe de varias formas, desde una lista, un diccionario, un array de NumPy o importando archivos. Para importar archivos de Excel es necesario instalar los siguientes paquetes:

```bash
pip install openpyxl
pip install xlrd
```

````python
# creacion desde listas

columnas = ['col1', 'col2', 'col3', 'col4']

lista = [[1,2,3,4],
         [5,6,7,8]]


df = pd.DataFrame(data=lista, columns=columnas)

df
````

````python
# desde numpy

import numpy as np

array = np.random.random((10, 4))

df = pd.DataFrame(data=array, columns=columnas)

df
````

````python
# desde dictio

dictio = {'a': [1,2,3,4,5],
          'b': [5,6,7,8,9]}


df = pd.DataFrame(data=dictio)

df
````

## 5 - Filtrado

Podemos filtrar el dataframe según los datos que este tiene en su interior. Podemos filtrar por una columna o por varias columnas al mismo tiempo. En este caso, ya no se usan las palabras `and` y `or`, sino los operadores binarios `&` y `|` respectivamente.

````python
df[df['neighbourhood_group_cleansed']=='Latina'].head()
````

````python
df[df['price']<100].head()
````

````python
df[(df['neighbourhood_group_cleansed']=='Latina') & (df['price']<100)].head()
````

````python
df[(df['neighbourhood_group_cleansed']=='Latina') | (df['price']<100)].head()
````

````python
df[(df['neighbourhood_group_cleansed']=='Latina') | ~(df['price']<100)].head()
````

## 6 - Agrupación

Con la tabla que estamos manejando, supongamos que queremos saber cuál es el precio medio del AirBnB en cada barrio. Para ver ese dato tenemos que agrupar por barrio toda la tabla para tener un barrio por cada fila y después calcular la media. Esa función de cálculo se le llama función de agregación y puede ser cualquier función, el máximo, el mínimo o la mediana. Para hacer esta operación de agrupación, pandas tiene el método `groupby`. Veamos como usarlo:

````python
df.groupby('neighbourhood_group_cleansed')
````

Hemos agrupado por barrio y nos devuelve un objeto de pandas. La pregunta es, ¿qué hacemos con el resto de columnas?, ¿qué función le aplicamos a cada una?. Usaremos el método `agg` de la agrupación para aplicar las funciones que queramos a las columnas que nos interesan.

````python
df.groupby('neighbourhood_group_cleansed').agg({'price': 'mean'})
````

Podemos agrupar por varias columnas a la vez y también aplicar distintas funciones de agregación a distintas columnas.

````python
df.groupby(['neighbourhood_group_cleansed', 'room_type']).agg({'price': 'mean'})
````

````python
df.groupby('neighbourhood_group_cleansed').agg({'price': 'mean', 'security_deposit': 'max'})
````

## 7 - Valores nulos

En pandas, los valores nulos, o faltantes, son aquellos que están ausentes o no definidos. Estos valores se representan comúnmente como `NaN` (Not a Number) en los DataFrames y Series de pandas. La gestión de valores nulos es una tarea común y esencial en el análisis de datos, ya que los datos faltantes pueden afectar los resultados de los análisis y modelos. Hay que tener en cuenta que un valor podría tener un significado. Sea como sea es necesario eliminarlos, ya sea borrando o rellenando dichos valores.

````python
lst_lst = [[1,2,3],
          ['3', 4, 5],
          ['3', 4]]


columns =['a', 'b', 'c']


df_lst = pd.DataFrame(data=lst_lst, columns=columns)

df_lst
````

````python
df_lst.fillna(0, inplace=True)

df_lst
````

## 9 - Aplicación de funciones

Podemos aplicar funciones propias al dataframe con el método `apply`. Este método es un manera de evitar bucles for en un dataframe, la función se aplica sobre la columna que deseemos de manera más eficiente. Su funcionamiento es igual que la función `map`.

````python
def cuadrado(x):
    
    return x**2



df['price'].apply(cuadrado).head()
````

````python
df['price'].apply(lambda x: x**2).head()
````
