## 1 - Modelos de texto a audio

Los modelos de texto a audio (Text-to-Audio) son sistemas que convierten texto escrito en sonido, ya sea en forma de habla humana, música o sonidos ambientales. Estos modelos utilizan técnicas avanzadas de inteligencia artificial y aprendizaje profundo para generar audio a partir de descripciones textuales. 


**Componentes principales de los modelos de texto a audio**


1. **Preprocesamiento de texto**:

    + Tokenización: Dividir el texto en unidades más pequeñas, como palabras o subpalabras.
    + Análisis Prosódico: Determinar la entonación, el ritmo y el énfasis adecuados para el texto.


2. **Modelo de conversión de texto a características intermedias**:

    + Codificador: Transforma el texto en una representación intermedia, como un espectrograma o una secuencia de características acústicas.
    + Redes Neuronales: Utilizan arquitecturas como RNNs, Transformers o Tacotron para mapear el texto a estas características intermedias.


3. **Síntesis de audio**:

    + Vocoders: Herramientas como WaveNet, Griffin-Lim o HiFi-GAN que convierten las características intermedias en señales de audio.


<br>


**Tipos de modelos de texto a audio**


1. **Modelos de texto a voz (Text-to-Speech, TTS)**:

    + Tacotron: Un modelo que convierte texto en espectrogramas que luego son transformados en audio mediante vocoders como WaveNet.
    + WaveNet: Un vocoder que genera señales de audio de alta calidad directamente a partir de espectrogramas.
    + Transformer TTS: Modelos como FastSpeech que utilizan arquitecturas de transformadores para mejorar la eficiencia y la velocidad de la síntesis de voz.


2. **Modelos de texto a música**:

    + Jukedeck: Utiliza redes neuronales para generar música a partir de descripciones textuales.
    + OpenAI Jukebox: Un modelo que genera música completa, incluyendo voz y letras, a partir de descripciones textuales.


3. **Modelos de texto a sonido ambiental**:

    + GANSynth: Utiliza Generative Adversarial Networks (GANs) para generar sonidos musicales y ambientales a partir de descripciones textuales.
    + Soundraw: Genera música y efectos de sonido basados en descripciones textuales, utilizando una combinación de técnicas de síntesis de audio y generación de texto a audio.

<br>

**Ejemplos de uso**

+ Asistentes virtuales: Utilizan TTS para responder a consultas de los usuarios con una voz natural.
+ Generación de contenido multimedia: Creación de música y efectos de sonido para videojuegos, películas y otros medios digitales.
+ Accesibilidad: Proveer tecnologías de lectura para personas con discapacidades visuales.

<br>

**Desafíos y avances recientes**

+ Naturalidad y expresividad: Mejorar la naturalidad y la expresividad de las voces generadas sigue siendo un desafío.
+ Eficiencia y velocidad: Modelos como FastSpeech y WaveGlow están diseñados para ser más rápidos y eficientes, permitiendo la generación de voz en tiempo real.
+ Adaptabilidad: Capacidad de adaptar modelos a diferentes voces, estilos de habla y sonidos específicos.


## 2 - Pipeline para modelos texto a audio

Usaremos el modelo [MusicGen de Facebook](https://huggingface.co/facebook/musicgen-medium) el cual viene en formato small de 2.4Gb, medium de 8Gb y large de 14Gb. También existe en versión estéreo. Este modelo está desarrollado por el equipo de investigación de IA de Meta, es una innovadora herramienta para la generación de música a partir de descripciones textuales. 


**Características principales**


1. **Arquitectura avanzada**:

    + Modelo Auto-regresivo: MusicGen utiliza un modelo Transformer auto-regresivo en una sola etapa. Este enfoque permite una generación de música eficiente y precisa, capturando las sutilezas y complejidades de la composición musical.
    + Tokenización de audio: Emplea un modelo EnCodec para la tokenización de audio, lo que facilita la transformación del texto en representaciones discretas de audio. Este modelo genera cuatro flujos melódicos paralelos que representan diferentes elementos musicales como bajos, percusión y armonía.


2. Variedad de tamaños de modelo: MusicGen está disponible en varios tamaños de modelo: small (300M - 2.4Gb), medium (1.5B - 8Gb) y large (3.3B - 14Gb) y una variante optimizada para la generación guiada por melodía. Esto permite a los usuarios elegir el modelo que mejor se adapte a sus necesidades y recursos computacionales.


3. Calidad de las muestras de música: Las muestras generadas por MusicGen se destacan por su claridad y riqueza, logrando imitar la profundidad y emoción de la música creada por artistas humanos. Esto marca un avance significativo en la calidad de la música generada por IA.


<br>

**Usos y aplicaciones**


1. Composición musical: Asistir a músicos en la creación de nuevas composiciones transformando descripciones textuales en música. Superar bloqueos creativos generando ideas melódicas o armónicas basadas en descripciones textuales.


2. Generación de banda sonora: Crear bandas sonoras únicas para películas, videojuegos y otras aplicaciones multimedia.


3. Herramientas educativas: Ayudar en la educación musical, especialmente en la comprensión de la composición y estructura musical.


4. Creación personalizada de música: Generar música personalizada basada en preferencias individuales o estados de ánimo.

<br>

**Limitaciones del Modelo**


+ Variabilidad en estilos musicales: El rendimiento del modelo puede variar entre diferentes estilos y géneros musicales, debido a la diversidad y complejidad inherente de estos estilos.


+ Profundidad emocional: Aunque el modelo genera música de alta calidad, puede enfrentar desafíos para capturar la profundidad emocional y la sutileza que un compositor humano aporta a la música.


+ Entrenamiento y adaptabilidad: MusicGen fue entrenado principalmente con descripciones en inglés, lo que puede limitar su desempeño en otros idiomas y culturas musicales.


````python
# para quitar warnings 

from transformers import logging
logging.set_verbosity_error()

import warnings
warnings.filterwarnings('ignore')
````

````python
# importamos desde la librería transformers el pipeline

from transformers import pipeline
````

````python
# definimos la tarea y el modelo

tarea = 'text-to-audio'  

modelo = 'facebook/musicgen-medium'
````

````python
# iniciamos el modelo de generación de audio, los modelos se descargan en local, en este caso son unos 8Gb

audio_pipe = pipeline(task=tarea, model=modelo)
````

````python
# descripoción de la música

prompt = 'jazz music with a piano melody, bass and drums.'
````

````python
# parámetros para ajustar la salida


params = {'temperature': 0.7,      # cuanto más alto más "creatividad" (aleatorio)
          'max_new_tokens': 150    # cuanto más alto más larga la respuesta
         }
````

````python
%%time

# llamada al pipeline con texto, la respuesta es un array de NumPy y la frecuencia de muestreo

respuesta = audio_pipe(text_inputs=prompt,
                       generate_kwargs=params)

respuesta
````

````python
# escuchamos el audio

from IPython.display import Audio

Audio(data=respuesta['audio'][0][0], rate=respuesta['sampling_rate'])
````

````python
# podemos guardar el audio en formato wav con scipy

import scipy

scipy.io.wavfile.write(filename='../../../files/jazz_sample.wav', 
                       rate=respuesta['sampling_rate'], 
                       data=respuesta['audio'][0][0])
````

## 3 - Uso del modelo

````python
# importamos los objetos procesador y modelo

from transformers import AutoProcessor, MusicgenForConditionalGeneration
````

### 3.1 - Procesador

El AutoProcessor es una clase en la biblioteca transformers de Hugging Face, diseñada para facilitar la carga y el uso de procesadores específicos para diferentes modelos de procesamiento de lenguaje natural y audio. Es una herramienta que permite manejar automáticamente diversos tipos de preprocesamiento y postprocesamiento de datos, dependiendo del tipo de modelo que se esté utilizando. Ya hemos utilizado este objeto.

````python
# inicialización del procesador

procesador = AutoProcessor.from_pretrained(pretrained_model_name_or_path='facebook/musicgen-medium')
````

**Descripción del procesador**

El `MusicgenProcessor` y su componente `EncodecFeatureExtractor` trabajan juntos para preparar los datos de audio de manera que sean utilizables por el modelo MusicGen. Este preprocesamiento incluye la normalización del audio, la segmentación en fragmentos manejables, y la generación de características que capturan la esencia del contenido del audio. Esto asegura que el modelo pueda aprender y generar música de manera efectiva y eficiente.


```
MusicgenProcessor:
- feature_extractor: EncodecFeatureExtractor {
  "chunk_length_s": null,
  "feature_extractor_type": "EncodecFeatureExtractor",
  "feature_size": 1,
  "overlap": null,
  "padding_side": "left",
  "padding_value": 0.0,
  "processor_class": "MusicgenProcessor",
  "return_attention_mask": true,
  "sampling_rate": 32000
}
```



1. **feature_extractor: EncodecFeatureExtractor.** Es un extractor de características diseñado para trabajar con el modelo MusicGen. Su función principal es convertir las señales de audio crudas en características que el modelo puede utilizar para generar música.


2. **chunk_length_s: null.** Define la longitud de los segmentos de audio en segundos. En este caso, es null, lo que significa que no se especifica una longitud de segmento fija y se puede procesar el audio completo de una sola vez o en fragmentos determinados por otros parámetros.


3. **feature_extractor_type: "EncodecFeatureExtractor".** Indica el tipo de extractor de características que se utiliza. En este caso, es un EncodecFeatureExtractor, que está optimizado para trabajar con datos de audio en el contexto del modelo MusicGen.


4. **feature_size: 1.** Define el tamaño de las características extraídas. Un valor de 1 puede indicar que se está utilizando una característica específica por muestra de audio, aunque en algunos contextos podría ser más complejo.


5. **overlap: null.** Especifica el solapamiento entre los segmentos de audio procesados. En este caso, null significa que no se está aplicando ningún solapamiento explícito entre los fragmentos de audio.


6. **padding_side: "left".** Indica que cualquier padding (relleno) añadido a las secuencias de audio se hará en el lado izquierdo. Esto es útil para alinear secuencias de diferentes longitudes para procesamiento en lotes.


7. **padding_value: 0.0.** El valor de padding utilizado es 0.0. Esto significa que las áreas de la secuencia que requieren padding se rellenarán con ceros, lo que es común en el procesamiento de audio para evitar introducir ruido adicional.


8. **processor_class: "MusicgenProcessor".** Especifica que esta configuración pertenece a la clase MusicgenProcessor, que maneja el preprocesamiento específico para el modelo MusicGen.


9. **return_attention_mask: true.** Indica que el extractor de características devolverá una máscara de atención. Esta máscara es útil para indicar qué partes de la secuencia son relevantes (no padding) y deben ser procesadas por el modelo.


10. **sampling_rate: 32000.** Define la tasa de muestreo del audio en 32000 Hz (32 kHz). Esta es la frecuencia a la cual se digitaliza el audio, y es importante para mantener la calidad y la coherencia de las características extraídas.


````python
# procesamiento del texto, podemos pasar varios

tensor = procesador(text=['80s pop track with bassy drums and synth', 
                          '90s rock song with loud guitars and heavy drums'],
                    
                    padding=True,
                    return_tensors='pt',
                   )

tensor
````

### 3.2 - Modelo MusicgenForConditionalGeneration

MusicgenForConditionalGeneration es una clase de la biblioteca transformers de Hugging Face, específicamente diseñada para la generación condicional de música utilizando el modelo MusicGen desarrollado por Meta AI (Facebook). Esta clase permite a los usuarios generar música basada en ciertas condiciones o descripciones textuales, integrando técnicas avanzadas de procesamiento de lenguaje natural y síntesis de audio.

````python
# inicialización del modelo

modelo = MusicgenForConditionalGeneration.from_pretrained(pretrained_model_name_or_path='facebook/musicgen-medium')
````

**Descripción del modelo**


1. **Text Encoder (codificador de texto)**:

    + T5EncoderModel: Utiliza el modelo T5 para codificar el texto de entrada.
    + Embedding: Matriz de embedding compartida para transformar las palabras en vectores de 768 dimensiones.
    + T5Stack: Conjunto de capas de bloques T5 que procesan estos embeddings.
    + T5Block: Bloques que comprenden capas de autoatención (T5LayerSelfAttention) y capas de feed-forward (T5LayerFF).
    + SelfAttention: Mecanismo de atención dentro de cada bloque T5, con proyecciones lineales para las consultas (q), claves (k), y valores (v).
    + LayerNorm: Normalización de las capas para estabilizar el entrenamiento.
    + Dropout: Mecanismo de regularización para prevenir el sobreajuste.


2. Audio Encoder (codificador de audio):

    + EncodecModel: Modelo utilizado para extraer características del audio.
    + EncodecEncoder: Codifica la señal de audio en una representación intermedia.
    + Conv1d: Convoluciones en una dimensión para procesar el audio.
    + ResnetBlock: Bloques residuales que mejoran la capacidad de aprendizaje del modelo.
    + LSTM: Capas de LSTM para capturar dependencias temporales en el audio.
    + EncodecDecoder: Decodifica la representación intermedia de vuelta a audio.
    + Quantizer: Cuantizador residual vectorial para compresión y codificación del audio.


3. Decoder (decodificador):

    + MusicgenForCausalLM: Modelo para la generación causal de música.
    + MusicgenModel: Modelo principal que comprende el decodificador.
    + MusicgenDecoder: Decodificador con capas de atención y capas feed-forward.
    + MusicgenAttention: Mecanismo de atención en el decodificador.
    + LayerNorm: Normalización de las capas.
    + Linear: Capas lineales para transformaciones de los datos.
    + lm_heads: Capas de salida lineales que transforman la representación interna en características finales de audio.


4. Proyección del codificador al decodificador:

    + enc_to_dec_proj: Proyección lineal que ajusta las dimensiones de salida del codificador de texto a las dimensiones de entrada del decodificador de audio.



<br>

**Funcionamiento del modelo**


1. Entrada de texto: Se proporciona una descripción textual de la música deseada, como "una melodía alegre con guitarra y percusión".


2. Codificación del texto: El texto se codifica utilizando el T5EncoderModel, que transforma la descripción en una representación vectorial.


3. Entrada de audio (Opcional): Si se proporciona un audio de referencia, el EncodecModel lo codifica en características intermedias. Para esto, se puede usar el modelo `musicgen-melody`, el cual necesita otro sistema operativo distinto de MacOSX. También el modelo [Stable Audio](https://huggingface.co/stabilityai/stable-audio-open-1.0) puede funcionar así.


4. Proyección: Las características codificadas del texto se proyectan a la dimensión esperada por el decodificador de audio.


5. Generación de música: El MusicgenForCausalLM utiliza las representaciones proyectadas para generar una secuencia de audio que cumple con la descripción textual proporcionada.


````python
%%time

# respuesta del modelo

respuesta = modelo.generate(**tensor, max_new_tokens=150)

respuesta
````
