## 1 - Modelos de texto a voz

Los modelos de texto a voz (Text-to-Speech, TTS) son sistemas que convierten texto escrito en habla. Estos modelos utilizan diversas técnicas de NLP y síntesis de voz para generar audio que suene natural.


**Componentes principales**


1. **Preprocesamiento del texto**:

    + Normalización: Transformar el texto a un formato estándar, por ejemplo expandir abreviaturas.
    + Tokenización: Dividir el texto en unidades más pequeñas, como palabras o caracteres.
    + Análisis prosódico: Determinar la entonación, el ritmo y el énfasis adecuado para el texto.


2. **Modelo de conversión de texto a voz**:

    + Encoder: Convierte el texto preprocesado en una representación intermedia.
    + Decoder: Transforma la representación intermedia en una secuencia de espectrogramas o directamente en audio.


3. **Síntesis de voz**:

    + Vocoders: Herramientas que convierten los espectrogramas generados en señales de audio. Ejemplos incluyen WaveNet, Griffin-Lim, y WaveGlow.

<br>

**Técnicas y modelos comunes**


1. **Modelos basados en concatenación**: Estos modelos utilizan fragmentos grabados de habla humana y los concatenan para formar palabras y frases. Aunque pueden producir audio de alta calidad, están limitados en términos de flexibilidad y naturalidad en diferentes contextos.


2. **Modelos paramétricos**: Utilizan parámetros para modelar las características de la voz humana como la frecuencia fundamental o el timbre. Los modelos Hidden Markov Models (HMM) fueron comunes en esta categoría antes del advenimiento de las redes neuronales profundas.


3. **Modelos basados en redes neuronales**:

    + Tacotron: Un modelo que convierte texto en espectrogramas, que luego son convertidos en audio mediante un vocoder como WaveNet o Griffin-Lim.
    + WaveNet: Un vocoder basado en redes neuronales profundas que genera señales de audio directamente a partir de espectrogramas, produciendo voces muy naturales.
    + Transformer TTS: Modelos como FastSpeech que utilizan arquitecturas de transformadores para mejorar la eficiencia y la velocidad de la síntesis de voz.


4. **Modelos de aprendizaje profundo**:

    + DeepVoice: Serie de modelos desarrollados por Baidu que mejoran la calidad de la voz sintética utilizando redes neuronales profundas.
    + WaveGlow: Un modelo de vocoder que combina flujos normalizantes y WaveNet para generar audio de alta calidad de manera eficiente.

<br>


**Desafíos y avances recientes**


1. **Naturalidad y expresividad**: Lograr que las voces sintéticas suenen más naturales y expresivas sigue siendo un desafío. Los modelos modernos están incorporando mejor manejo de prosodia y emociones.


2. **Velocidad y eficiencia**: Modelos como FastSpeech y WaveGlow están diseñados para ser más rápidos y eficientes, permitiendo la generación de voz en tiempo real.


3. **Adaptabilidad**: La capacidad de adaptar modelos a diferentes voces y estilos de habla, incluyendo voces multi-idioma y con diferentes acentos, está mejorando.


4. **Accesibilidad y personalización**: Los modelos TTS están siendo utilizados para crear soluciones más accesibles para personas con discapacidades visuales y para personalizar experiencias de usuario en aplicaciones y dispositivos.

<br>

**Aplicaciones**

+ Asistentes virtuales: Siri, Alexa o Google Assistant.
+ Lectores de pantalla: Para personas con discapacidades visuales.
+ Generación de contenido: Audiolibros, doblaje automático de videos.
+ Comunicación aumentativa y alternativa (CAA): Ayudando a personas con trastornos del habla a comunicarse.

## 2 - Pipeline para modelos texto a voz

Vamos a usar el modelo [facebook/mms-tts](https://huggingface.co/facebook/mms-tts-spa) en castellano que tiene un peso aproximado de 150Mb. Es parte del proyecto Massively Multilingual Speech (MMS) de Facebook, diseñado para proporcionar tecnología de síntesis de voz en una amplia gama de idiomas. 


**Características del modelo**


1. **Arquitectura**:

    + VITS: Utiliza la arquitectura VITS (Variational Inference with adversarial learning for end-to-end Text-to-Speech), que es un modelo de síntesis de voz de extremo a extremo. Este modelo predice una forma de onda del habla a partir de una secuencia de texto de entrada.
    + Autoencoder variacional: Consiste en un codificador posterior, un decodificador y un previo condicional. Utiliza un módulo basado en flujo y un codificador de texto basado en Transformer para predecir características acústicas espectrales.


2. **Capacidad multilingüe**:

    + Cobertura de idiomas: El proyecto MMS se enfoca en expandir la tecnología de habla a más de 1000 idiomas, utilizando datos de múltiples fuentes para entrenar modelos que pueden manejar una gran diversidad lingüística.
    + Adaptación a diferentes idiomas: Cada idioma tiene su propio punto de control VITS, entrenado específicamente para las características de ese idioma.


3. **Entrenamiento y generación**:

    + Adversarial training y flujos de normalización: El modelo se entrena con una combinación de pérdidas derivadas de la cota inferior variacional y el entrenamiento adversarial. Los flujos de normalización se aplican a la distribución prior condicional para mejorar la expresividad.
    + Predicción estocástica de duración: Incluye un predictor de duración estocástica que permite sintetizar el habla con diferentes ritmos a partir del mismo texto de entrada, lo que refleja la naturaleza no determinista del habla humana.


```python
# para quitar warnings 

from transformers import logging
logging.set_verbosity_error()

import warnings
warnings.filterwarnings('ignore')
```

```python
# importamos desde la librería transformers el pipeline

from transformers import pipeline
```

```python
# definimos la tarea y el modelo

tarea = 'text-to-speech'  

modelo = 'facebook/mms-tts-spa'
```

```python
# iniciamos el modelo TTS, los modelos se descargan en local, en este caso son unos 150Mb

tts_pipe = pipeline(task=tarea, model=modelo)
```

```python
# llamada al pipeline con texto, la respuesta es un array de NumPy y la frecuencia de muestreo

respuesta = tts_pipe('Hola')

respuesta
```

```python
# escuchamos el audio

from IPython.display import Audio

Audio(respuesta['audio'], rate=respuesta['sampling_rate'])
```

```python
# Ejemplo Unamuno


texto = '''
        Cuando alguien te pida algo, no le des lo que te pide.
        Dale lo que tú consideras que necesita y soporta después 
        la ingratitud.
        '''


respuesta = tts_pipe(texto)
```

```python
# escuchamos el audio

Audio(respuesta['audio'], rate=respuesta['sampling_rate'])
```

### 3.1 - Tokenizador

El AutoTokenizer es una clase de la biblioteca transformers diseñada para facilitar la creación y uso de tokenizadores específicos para modelos NLP. Ya hemos usado varias veces este objeto. Ahora vamos a cargar el tokenizador TTS.

```python
tokenizador = AutoTokenizer.from_pretrained('facebook/mms-tts-spa')

tokenizador
```

**Descripción del VitsTokenizer**


+ `name_or_path='facebook/mms-tts-spa'`: Especifica el nombre o la ruta del modelo preentrenado que está utilizando el tokenizador. En este caso, el modelo de TTS para el español (spa).


+ `vocab_size=45`: Indica que el tamaño del vocabulario es de 45 tokens. Este es el número de tokens únicos que el modelo puede reconocer y procesar.


+ `model_max_length=1000000000000000019884624838656`: Define la longitud máxima del modelo. Este número extremadamente grande sugiere que, en la práctica, no hay una longitud máxima efectiva que el tokenizador imponga en las secuencias de entrada.


+ `is_fast=False`: Indica que no se está utilizando la versión rápida del tokenizador. Los tokenizadores rápidos son implementaciones más eficientes que utilizan Rust para acelerar el proceso de tokenización.


+ `padding_side='right' y truncation_side='right'`: Especifica que el padding (relleno) y la truncación se aplican al lado derecho de las secuencias de tokens. Esto significa que cualquier espacio adicional necesario para cumplir con una longitud fija se agregará al final de la secuencia.


+ `special_tokens={'unk_token': '<unk>', 'pad_token': '7'}`: Define los tokens especiales utilizados por el tokenizador. `<unk>` (unknown token) se utiliza para representar cualquier token que no esté en el vocabulario. 7 (padding token) se utiliza para el relleno en las secuencias para mantener una longitud uniforme.


+ `clean_up_tokenization_spaces=True`: Indica que el tokenizador limpiará los espacios en blanco adicionales después de la tokenización. Esto ayuda a mantener las secuencias de tokens más ordenadas.




```python
# texto a tokenizar

texto = 'hola que tal como estas'
```

```python
# tokenizar texto

tensor = tokenizador(text=texto, return_tensors='pt')

tensor
```

### 3.2 - Modelo Vits

El VitsModel, basado en este [paper](https://arxiv.org/abs/2106.06103), es un modelo de síntesis de voz de extremo a extremo que convierte texto en habla, utilizando una combinación de inferencia variacional y aprendizaje adversarial. 


**Características principales del VitsModel**


1. **Inferencia variacional con aprendizaje adversarial**: VITS combina técnicas de inferencia variacional con aprendizaje adversarial para generar formas de onda del habla a partir de secuencias de texto. Este enfoque permite una síntesis de voz más natural y expresiva.


2. **Arquitectura del modelo**:

    + Codificador-Decoder: Utiliza un codificador para transformar el texto en una representación intermedia y un decodificador para convertir esta representación en un espectrograma.
    + Autoencoder variacional (VAE): El modelo incluye un autoencoder variacional que ayuda a capturar la variabilidad en la síntesis de voz, permitiendo la generación de diferentes estilos y ritmos del habla a partir del mismo texto.
    + Módulo basado en flujos: Predice características acústicas espectrales utilizando un módulo basado en flujos, compuesto por un codificador de texto basado en Transformer y múltiples capas de acoplamiento.
    + HiFi-GAN Vocoder: Utiliza un vocoder HiFi-GAN para decodificar el espectrograma en una forma de onda de audio final, produciendo una voz de alta fidelidad.


3. **Predicción de duración estocástica**: Incluye un predictor de duración estocástica que permite al modelo sintetizar habla con diferentes ritmos a partir del mismo texto de entrada. Esto refleja la naturaleza no determinista del habla humana.


4. **Entrenamiento y pérdidas**: El modelo se entrena de manera end-to-end utilizando una combinación de pérdidas derivadas de la cota inferior variacional y el entrenamiento adversarial. Esto mejora la calidad y naturalidad del habla generada.


**Descripción del VitsModel**


1. `Text Encoder (codificador de texto)`:

    + Embedding Layer: Embedding(45, 192) - Esta capa transforma los tokens de texto en vectores de 192 dimensiones.
    + Encoder: VitsEncoder - Un conjunto de capas que procesan los vectores de embedding.
    + Layers: ModuleList con 6 capas VitsEncoderLayer.
    + Attention: VitsAttention - Mecanismo de atención que permite al modelo enfocarse en partes específicas del texto.
    + Feed Forward: VitsFeedForward - Redes convolucionales de una dimensión que aplican una transformación no lineal a los datos.
    + Layer Normalization: LayerNorm - Normaliza la salida de las capas para estabilizar y acelerar el entrenamiento.


2. `Flow Module (módulo de flujo)`:

    + Residual Coupling Block: VitsResidualCouplingBlock - Un bloque que facilita el aprendizaje eficiente de transformaciones complejas.
    + Flows: ModuleList con 4 capas VitsResidualCouplingLayer.
    + WaveNet: VitsWaveNet - Utiliza convoluciones parametrizadas para modelar la distribución de las características acústicas.


3. `Decoder (decodificador)`:

    + HiFi-GAN: VitsHifiGan - Un vocoder avanzado que convierte espectrogramas en formas de onda de alta calidad.
    + Upsampler: ModuleList con capas ConvTranspose1d para aumentar la resolución temporal de los espectrogramas.
    + Residual Blocks: HifiGanResidualBlock - Bloques residuales que mejoran la capacidad del modelo para generar detalles finos en el audio.


4. `Duration Predictor (predictor de duración)`:

    + Stochastic Duration Predictor: VitsStochasticDurationPredictor - Predice la duración de cada fonema de manera estocástica, permitiendo variabilidad en la síntesis de voz.
    + Dilated Depth Separable Convs: VitsDilatedDepthSeparableConv - Convoluciones dilatadas que mejoran la capacidad del modelo para capturar dependencias temporales largas.


5. `Posterior Encoder (codificador posterior)`:

    + WaveNet: VitsWaveNet - Utilizado aquí para modelar la distribución posterior de las características acústicas del habla.
    + Parametrized Convs: Convoluciones parametrizadas con normalización de peso para mejorar la estabilidad del entrenamiento.


<br>

**Funcionamiento del VitsModel**


1. Entrada de texto: El texto de entrada se tokeniza y se transforma en vectores de embedding.


2. Codificación del texto: Los embeddings pasan a través de múltiples capas del VitsEncoder, donde se aplican mecanismos de atención y capas de feed-forward.


3. Módulo de flujo: El flujo residual y el WaveNet transforman los vectores de características intermedias en representaciones acústicas.


4. Predicción de duración: El Stochastic Duration Predictor predice la duración de cada fonema, añadiendo variabilidad a la síntesis de voz.


5. Decodificación: El HiFi-GAN toma los espectrogramas generados y los convierte en formas de onda de alta calidad, que representan el audio final.


6. Salida de audio: El resultado es una forma de onda de audio que suena natural y es coherente con el texto de entrada.
