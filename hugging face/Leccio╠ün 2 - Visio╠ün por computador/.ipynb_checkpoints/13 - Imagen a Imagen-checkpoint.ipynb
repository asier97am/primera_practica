## 1 - Modelos de imagen a imagen

Los modelos de imagen a imagen son una categoría de algoritmos que se utilizan para transformar una imagen de una representación a otra. Estos modelos son capaces de realizar tareas complejas de procesamiento de imágenes, incluyendo la modificación, generación y mejora de imágenes. Estos modelos de imagen a imagen han revolucionado el campo del procesamiento de imágenes, permitiendo transformaciones visuales que antes eran imposibles o muy difíciles de lograr. Utilizan técnicas avanzadas de aprendizaje profundo y redes neuronales para aprender y generar representaciones visuales complejas, abriendo nuevas posibilidades en la creatividad digital, la mejora de imágenes y la traducción de estilos visuales. 



**Descripción de los Modelos de Imagen a Imagen**


1. **Transformación de imagen a imagen**: Los modelos de imagen a imagen se entrenan para mapear una imagen de un dominio de entrada a una imagen correspondiente en un dominio de salida. Utilizan técnicas de aprendizaje profundo y redes neuronales, específicamente redes neuronales convolucionales (CNN) y redes generativas antagónicas (GAN), para aprender a realizar estas transformaciones a partir de conjuntos de datos de entrenamiento.


2. **Redes Generativas Antagónicas (GAN)**: Las GAN consisten en dos redes, una generativa que crea imágenes falsas a partir de ruido o de una entrada específica, y una discriminativa que intenta distinguir entre imágenes reales y generadas. A través de un proceso de entrenamiento conjunto, la red generativa mejora en la producción de imágenes que la red discriminativa no puede diferenciar de las reales. Las GAN se utilizan ampliamente en modelos de imagen a imagen para generar imágenes de alta calidad y realismo.




**Algunos modelos**:


1. **Pix2Pix**: Un modelo basado en redes GAN que fue diseñado para aprender mapeos condicionales de imagen a imagen. Se entrena con pares de imágenes que tienen una correspondencia directa entre sí. Puede ser utilizado para tareas como traducir bocetos a fotografías, convertir imágenes en blanco y negro a color, y transformar mapas de segmentación en imágenes fotográficas.


2. **CycleGAN**: También se basa en GANs, pero a diferencia de Pix2Pix, no requiere pares de imágenes correspondientes para entrenarse. Utiliza dos mapeos, de dominio A a dominio B y viceversa, y emplea una técnica llamada "consistencia cíclica" para asegurar que una imagen transformada vuelva a su forma original cuando se aplique la transformación inversa. Es útil para tareas donde no se disponen de pares de imágenes, como la traducción de estilos artísticos, por ejemplo, de Monet a fotografías modernas, la conversión de fotos de verano a invierno, y la transformación de caballos en cebras.


3. **SRGAN (Super-Resolution GAN)**: Está diseñado para mejorar la resolución de las imágenes. Utiliza una red generativa para aumentar la resolución de las imágenes de baja resolución, y una red discriminativa para distinguir entre imágenes de alta resolución generadas y reales. Mejora la calidad de imágenes en aplicaciones como la televisión de alta definición, imágenes médicas, y fotografías de satélite.


4. **StyleGAN**: Desarrollado por NVIDIA, es una variante de GAN que permite un control preciso sobre el estilo y las características de la imagen generada. Utiliza un concepto llamado "space disentanglement" para separar diferentes atributos de la imagen. Se puede usar para la generación de rostros humanos realistas, creación de arte digital y síntesis de nuevos estilos visuales a partir de ejemplos existentes.


5. **DeepDream**: Es un algoritmo desarrollado por Google que modifica imágenes para acentuar y exagerar patrones en ellas, utilizando redes neuronales convolucionales (CNN). Se puede usar para la generación de arte abstracto y visualización de lo que las redes neuronales han aprendido.


## 2 - Pipeline de Diffusers para modelos de imagen a imagen

Usemos el modelo [InstructPix2Pix](https://huggingface.co/timbrooks/instruct-pix2pix), que se entrena ajustando finamente un modelo StableDiffusion 1.5.

Si no estamos obteniendo la calidad de resultado que deseamos, puede haber varias razones. Por ejemplo, ¿la imagen no está cambiando lo suficiente?, el parámetro image_guidance_scale puede ser demasiado alto. Este valor dicta cuán similar debe ser la salida a la entrada. Alternativamente, el parámetroguidance_scale puede ser demasiado bajo. Este valor dicta cuánto debe escuchar la instrucción de texto. El image_guidance_scale por defecto de 1.5 y el guidance_scale de 7.5 son un buen punto de partida, pero no son necesariamente óptimos para cada edición. Podemos intentar disminuir el image_guidance_scale, o aumentar el guidance_scale.

Por el contrario, ¿la imagen está cambiando demasiado, de modo que los detalles en la imagen original no se conservan? Podemos intentar aumentar el image_guidance_scale o diminuir el guidance_scale. También reformular la instrucción a veces mejora los resultados, por ejemplo, "convertirlo en un robot" vs. "hacerlo un robot" vs. "como un robot") y aumentar el número de pasos suele mejorar los resultados. ¿Las caras se ven raras? El autoencoder de Stable Diffusion tiene dificultades con las caras que son pequeñas en la imagen. Intenta recortar la imagen para que la cara ocupe una mayor porción del encuadre.

````python
# para quitar warnings 

from transformers import logging
logging.set_verbosity_error()

import warnings
warnings.filterwarnings('ignore')
````

````python
# importamos desde la librería diffusers el pipeline

from diffusers import DiffusionPipeline
````

````python
# definimos el modelo

modelo = 'timbrooks/instruct-pix2pix' 
````

````python
# importamos torch y el scheduler

import torch
from diffusers import EulerAncestralDiscreteScheduler
````

````python
# iniciamos el modelo de difusión, los modelos se descargan en local, en este caso son unos 5.5Gb

pipe = DiffusionPipeline.from_pretrained(pretrained_model_name_or_path=modelo, 
                                         torch_dtype=torch.float16, 
                                         safety_checker=None)
````

````python
# cambio de dispositivo y añadido del scheduler al pipeline

pipe.to('mps')

pipe.scheduler = EulerAncestralDiscreteScheduler.from_config(pipe.scheduler.config)
````

Usaremos una imagen de referencia del modelo, el [David de Miguel Angel](https://raw.githubusercontent.com/timothybrooks/instruct-pix2pix/main/imgs/example.jpg).

````python
# url de la imagen

url = 'https://raw.githubusercontent.com/timothybrooks/instruct-pix2pix/main/imgs/example.jpg'
````

````python
# visualizamos la imagen con PIL en 400x400 píxeles

from PIL import Image
import requests as req

imagen = Image.open(req.get(url, stream=True).raw)

imagen.resize((400, 400))
````

````python
# texto de referencia

prompt = 'turn him into cyborg'
````

````python
# llamada al pipeline

resultado = pipe(prompt=prompt, 
                 image=imagen, 
                 num_inference_steps=50, 
                 image_guidance_scale=1, 
                 guidance_scale=7.5)

resultado
````

````python
# imagen modificada

resultado.images[0]
````

## 3 - Generador de QR

Vamos a usar un generador de QRs para luego darle un formato artístico con StableDiffusion. Primero tenemos que instalar la librería [qrcode](https://pypi.org/project/qrcode/) con el siguiente comando:

```bash
pip install qrcode
```

Vemos como se usa:

````python
# importamos la librería

import qrcode
````

````python
# creamos la imagen del qr con la url de una web

imagen_qr = qrcode.make('https://www.hackio.com/')
````

````python
# la guardamos y la visualizamos

imagen_qr.save('../../images/imagen_qr.png')

imagen_qr
````

Hemos guardado la imagen para luego cargarla como objeto PIL y pasársela al modelo de StableDiffusion. Usaremos el [StableDiffusion 2.1](https://huggingface.co/stabilityai/stable-diffusion-2-1) que ya hemos usado anteriormente. Además, para la generación del QR, usamos un modelo de control de [Vertelabs](https://huggingface.co/vertxlabs/controlnet_qrcode-control_v11p_v1). La principal ventaja de usar un modelo de control es la capacidad de guiar la red generativa para producir imágenes que sigan ciertas restricciones o características deseadas. Esto es particularmente útil en aplicaciones donde se necesita una precisión y control más altos sobre el resultado final. En este caso la idea es guiar al modelo para que cree un QR operativo. El modelo de control pesa unos 1.5Gb.

Este modelo funciona bastante bien en la mayoría de los casos, pero ten en cuenta que no es 100% precisos. En algunos casos, la forma del código QR podría no aparecer como se esperaba. Podemos aumentar el peso del modelo de control para enfatizar la forma del código QR. Sin embargo esto podría afectar negativamente el estilo del resultado.


````python
# modelo control

from diffusers import ControlNetModel

red_control = ControlNetModel.from_pretrained('vertxlabs/controlnet_qrcode-control_v11p_v1')
````

````python
# modelo de difusión

from diffusers import StableDiffusionControlNetPipeline

modelo = 'stabilityai/stable-diffusion-2-1'

pipe = StableDiffusionControlNetPipeline.from_pretrained(pretrained_model_name_or_path=modelo,
                                                         controlnet=red_control)
````

````python
# cambio de dispositivo y añadido del scheduler al pipeline

pipe.to('mps')

pipe.scheduler = EulerAncestralDiscreteScheduler.from_config(pipe.scheduler.config)
````

````python
# imagen del QR

imagen_qr = Image.open('../../images/imagen_qr.png')
````

````python
# textos

prompt = '''Blooming chinese chrysanthemum, 
            green leaves growing wantonly, flowers, 
            Complex patterns on the border,
            Masterpiece Art, Beauty, 8K,  Unreal Engine'''


negative_prompt = 'ugly, low quality, blurry, nsfw'
````

````python
# llamada al pipeline

resultado = pipe(prompt=prompt,
                 negative_prompt=negative_prompt, 
                 image=imagen_qr,
                 control_image=imagen_qr,
                 width=512,
                 height=512,
                 guidance_scale=20,
                 controlnet_conditioning_scale=2.,
                 strength=0.9, 
                 num_inference_steps=50,
                )
````

````python
resultado.images[0]
````

## 4 - Cartoonize

[Este modelo](https://huggingface.co/instruction-tuning-sd/cartoonizer) convierte una imagen a dibujos. Esta basado en un StableDiffusion 1.5 ajustado para convertir la imagen a dibujo. Tiene un peso aproximado de 5.5Gb.

````python
# importamos el pipeline

from diffusers import StableDiffusionInstructPix2PixPipeline
````

````python
# definimos el modelo

modelo = 'instruction-tuning-sd/cartoonizer'
````

````python
# definimos el pipeline

pipe = StableDiffusionInstructPix2PixPipeline.from_pretrained(pretrained_model_name_or_path=modelo,
                                                              torch_dtype=torch.float16)
````

````python
# cambio de dispositivo y añadido del scheduler al pipeline

pipe.to('mps')

pipe.scheduler = EulerAncestralDiscreteScheduler.from_config(pipe.scheduler.config)
````

Probemos la [imagen de una montaña](https://hf.co/datasets/diffusers/diffusers-images-docs/resolve/main/mountain.png).

````python
# url de la imagen 

url = 'https://hf.co/datasets/diffusers/diffusers-images-docs/resolve/main/mountain.png'
````

````python
# carga y visualización

imagen = Image.open(req.get(url, stream=True).raw)

imagen.resize((600, 400))
````

````python
# texto

prompt = 'Cartoonize the following image'
````

````python
# llamada al pipeline

resultado = pipe(prompt=prompt, image=imagen)
````

````python
# imagen final

resultado.images[0].resize((600, 400))
````

## 5 - Generación Incondicional

La generación incondicional de imágenes es un proceso que implica la creación de imágenes a partir de ruido aleatorio o un vector latente, sin ninguna condición adicional o entrada específica. Esto contrasta con la generación condicional de imágenes, donde la salida está condicionada por alguna forma de entrada, como texto, una imagen inicial, o algún otro tipo de información.

Veamos dos ejemplos, un modelo de Google que genera [iglesias](https://huggingface.co/google/ddpm-church-256) y otro modelo de Google que genera [gatos](https://huggingface.co/google/ddpm-ema-cat-256). Ambos pesan unos 550Mb.

````python
# importamos el pipeline

from diffusers import DDPMPipeline
````

````python
# definimos el modelo

modelo = 'google/ddpm-church-256'
````

````python
# definimos el pipeline

pipe = DDPMPipeline.from_pretrained(pretrained_model_name_or_path=modelo)
````

````python
# cambio de dispositivo 

pipe = pipe.to('mps')
````

````python
# llamada al pipeline

resultado = pipe(num_inference_steps=100)
````

````python
# imagen final

resultado.images[0].resize((400, 400))
````

````python
# modelo de gatos


# definimos el modelo
modelo = 'google/ddpm-ema-cat-256'


# definimos el pipeline
pipe = DDPMPipeline.from_pretrained(pretrained_model_name_or_path=modelo)


# cambio de dispositivo 
pipe = pipe.to('mps')


# llamada al pipeline
resultado = pipe(num_inference_steps=100)


# imagen final

resultado.images[0].resize((400, 400))
````
