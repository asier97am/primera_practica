## 1 - Modelos de imagen a texto

Los modelos de imagen a texto son sistemas diseñados para interpretar imágenes y generar descripciones textuales precisas de las mismas. Estos modelos combinan técnicas avanzadas de procesamiento de imágenes y generación de lenguaje natural para lograr sus objetivos. 


**Componentes de los Modelos de Imagen a Texto**


1. **Procesamiento de imágenes**: 

    + Redes convolucionales, CNNs, se utilizan para extraer características visuales de la imagen. Las CNNs son efectivas en la detección de patrones y objetos dentro de las imágenes, lo que permite al modelo entender el contenido visual.


2. **Codificación de características**: 

    + Las características extraídas por las CNNs se convierten en vectores que representan la información visual en un formato que puede ser interpretado por otros componentes del sistema.


3. **Generación de texto**:

    + Redes neuronales recurrentes, RNNs, y algunas variantes como LSTMs y GRUs, se utilizan para procesar las secuencias de características y generar descripciones textuales. Las LSTM (Long Short-Term Memory) y GRU (Gated Recurrent Unit) son variantes que manejan mejor las dependencias a largo plazo en las secuencias de datos.
    + Transformers: Un modelo más reciente que ha demostrado ser muy efectivo en la generación de texto. Los Transformers utilizan mecanismos de atención para gestionar mejor las relaciones dentro de la secuencia de datos, mejorando la coherencia y la fluidez del texto generado.


4. **Entrenamiento del modelo**:

    + Datos de Entrenamiento: El modelo se entrena con un conjunto de datos que contiene imágenes y sus descripciones correspondientes. Estos datos permiten al modelo aprender a asociar características visuales específicas con palabras y frases adecuadas.
    + Pérdida y Optimización: Se utiliza una función de pérdida para medir la diferencia entre las descripciones generadas y las reales durante el entrenamiento. Algoritmos de optimización como Adam o SGD se emplean para ajustar los pesos del modelo y minimizar la pérdida.



**Proceso de generación de texto a partir de una imagen**

1. Entrada de la imagen: El modelo recibe una imagen como entrada.

2. Extracción de características: La imagen se pasa a través de la CNN para extraer características visuales.

3. Codificación: Las características se codifican en un vector que representa el contenido visual de la imagen.

4. Decodificación: El vector codificado se pasa a través de una RNN, LSTM, GRU o Transformer para generar una secuencia de palabras que describen la imagen.

5. Salida de texto: El modelo produce una descripción textual de la imagen.


**Aplicaciones**

+ Descripciones de imágenes para personas con discapacidad visual: Ayuda a las personas con discapacidad visual a entender el contenido de las imágenes.

+ Etiquetado automático de fotos: Facilita la organización y búsqueda de fotos en bases de datos y redes sociales.

+ Asistentes virtuales y chatbots: Mejora la interacción de los asistentes virtuales y chatbots con usuarios mediante la interpretación de imágenes.

+ Generación de contenido: Automatiza la creación de descripciones para comercio electrónico, catálogos y medios digitales.

## 2 - Pipeline para modelos de imagen a texto

Usaremos el modelo [BLIP](https://huggingface.co/Salesforce/blip-image-captioning-large) (Bootstrapped Language-Image Pre-training) creado por Salesforce. Está diseñado para mejorar la comprensión y generación de tareas de visión y lenguaje. Utiliza una arquitectura basada en Transformers, que es altamente efectiva para tareas de secuencias de datos debido a su capacidad de manejar dependencias a largo plazo y relaciones contextuales complejas e incluye encoders para procesar las imágenes y decoders para generar descripciones textuales. Inicialmente preentrenado con grandes conjuntos de datos para aprender representaciones generales de imágenes y lenguaje y posteriormente ajustado con conjuntos de datos específicos para tareas como la creación de subtítulos de imágenes, mejorando su precisión y relevancia en esta tarea. El modelo tiene un peso aproximado de 2Gb.

```python
# importamos desde la librería transformers el pipeline

from transformers import pipeline
```

```python
# definimos la tarea y el modelo

tarea = 'image-to-text'  

modelo = 'Salesforce/blip-image-captioning-large'
```

```python
# iniciamos el descriptor de imagenes, los modelos se descargan en local, en este caso son unos 2Gb

img_pipe = pipeline(task=tarea, model=modelo, max_new_tokens=100)
```

```python
# llamada al pipeline con imagen en local

img_pipe('../../images/shinjuku.jpeg')
```

Usemos ahora una imagen de demostración del propio modelo, [una mujer sentada en la playa con su perro](https://storage.googleapis.com/sfr-vision-language-research/BLIP/demo.jpg).

```python
img_url = 'https://storage.googleapis.com/sfr-vision-language-research/BLIP/demo.jpg' 
```

```python
# visualizamos la imagen con PIL en 400x400 píxeles

from PIL import Image
import requests as req

imagen = Image.open(req.get(img_url, stream=True).raw)

imagen.resize((600, 400))
```

```python
# llamada al pipeline para la descripción de la imagen

img_pipe(imagen)
```

Otro ejemplo, la una imagen de [gatitos en un sofa](http://images.cocodataset.org/val2017/000000039769.jpg'). Usamos directamente la url.

```python
# otro ejemplo de uso

img_url = 'http://images.cocodataset.org/val2017/000000039769.jpg' 

img_pipe(img_url)
```

## 3 - Usando el modelo descriptor

```python
# importamos los objetos procesador y modelo descriptor de imagenes

from transformers import BlipProcessor, BlipForConditionalGeneration
```

### 3.1 - Procesador de imágenes y texto

BlipProcessor es una herramienta que forma parte del marco BLIP y está diseñada para facilitar el procesamiento y la preparación de datos tanto visuales como textuales para su uso en modelos de imagen y texto. 

**Funciones del BlipProcessor**

1. Procesamiento de Imágenes:

    + Redimensionamiento y normalización: Ajusta las imágenes a un tamaño adecuado y normaliza los valores de píxeles para que sean compatibles con el modelo.
    + Transformaciones: Aplica transformaciones como recorte, rotación, y cambio de escala para aumentar la robustez del modelo frente a variaciones en los datos de entrada.


2. Procesamiento de Texto:

    + Tokenización: Convierte el texto en una secuencia de tokens que pueden ser procesados por el modelo.
    + Codificación: Transforma los tokens en representaciones numéricas que pueden ser interpretadas por el modelo de lenguaje.


3. Integración de Datos Multimodales:

    + Fusión de imágenes y textos: Combina las representaciones de imágenes y textos en un formato que permite al modelo de visión y lenguaje entender y procesar ambos tipos de datos simultáneamente.


```python
# transforma las imágenes y el texto

procesador = BlipProcessor.from_pretrained('Salesforce/blip-image-captioning-large')
```

```python
# descripcion del objeto

procesador
```

**Componentes del BlipProcessor**


**BlipImageProcessor**

Este componente se encarga del preprocesamiento de imágenes, asegurando que estén en un formato adecuado para ser interpretadas por el modelo. 


+ **_valid_processor_keys**: Claves válidas que el procesador reconoce y utiliza.

    + **images**: Imágenes a procesar.
    + **do_resize**: Indica si las imágenes deben redimensionarse.
    + **size**: Tamaño al que se redimensionarán las imágenes, 384x384 píxeles.
    + **resample**: Método de remuestreo, 3 se refiere a Image.BILINEAR.
    + **do_rescale**: Indica si las imágenes deben reescalar.
    + **rescale_factor**: Factor de reescalado, 1/255.
    + **do_normalize**: Indica si las imágenes deben normalizarse.
    + **image_mean**: Media de los canales RGB para la normalización.
    + **image_std**: Desviación estándar de los canales RGB para la normalización.
    + **do_convert_rgb**: Convierte las imágenes a RGB si es necesario.
    + **return_tensors**: Formato de retorno de los tensores.
    + **data_format**: Formato de los datos de entrada.
    + **input_data_format**: Formato esperado de los datos de entrada.


+ **image_processor_type**: Tipo del procesador de imágenes.

+ **processor_class**: Clase del procesador.

+ **size_divisor**: Divisor para ajustar el tamaño, usualmente para asegurar compatibilidad con el tamaño del modelo.



**BertTokenizerFast**

Este componente se encarga del procesamiento y tokenización de texto, adaptando las entradas textuales al formato requerido por el modelo de lenguaje.

+ **name_or_path**: Ruta o nombre del modelo preentrenado.

+ **vocab_size**: 30522, tamaño del vocabulario.

+ **model_max_length**: 512, longitud máxima de los modelos.

+ **is_fast**: true, usa una versión rápida del tokenizador.

+ **padding_side**: 'right', lado del padding.

+ **truncation_side**: 'right', lado de la truncación.

+ **special_tokens**: Tokens especiales utilizados por el modelo.

    + unk_token: `[UNK]` Usado para tokens desconocidos.
    + sep_token: `[SEP]` Usado para separación de secuencias.
    + pad_token: `[PAD]` Usado para padding.
    + cls_token: `[CLS]` Usado para la clasificación.
    + mask_token: `[MASK]` Usado para enmascarar.

+ **clean_up_tokenization_spaces**: true, limpia los espacios en la tokenización.


```python
# transformación de la imagen y el texto, al añadir este texto se pone una condición

img_txt_procesado = procesador(images=imagen, text='a photography of', return_tensors='pt')
```

```python
# una imagen, 3 canales, 384x384 píxeles

img_txt_procesado['pixel_values'].shape
```

```python
# tokenizacion texto

img_txt_procesado['input_ids']
```

### 3.2 - Modelo descriptor de imágenes

BlipForConditionalGeneration es una clase de modelo dentro de marco BLIP diseñada para tareas de generación condicional. Esto significa que el modelo genera texto basado en una condición, que en este contexto generalmente es una imagen. Este tipo de modelo es útil para aplicaciones como la generación de subtítulos para imágenes, la respuesta a preguntas visuales (VQA) y otras tareas que requieren la generación de lenguaje basado en información visual.


**Características Clave de BlipForConditionalGeneration**

1. **Generación condicional**: El modelo genera texto, subtítulos, respuestas o descripciones, en función de una entrada condicional, que suele ser una imagen.


2. **Arquitectura basada en Transformers**: Utiliza una arquitectura de Transformer, que es altamente efectiva para manejar tareas secuenciales y de generación de texto debido a su capacidad de manejar relaciones contextuales complejas y dependencias a largo plazo.


3. **Entrenamiento multimodal**: Está preentrenado con datos que incluyen pares de imagen-texto, lo que le permite comprender y generar lenguaje basado en contenido visual de manera efectiva.


```python
# inicializacion del modelo Blip

modelo_descriptor = BlipForConditionalGeneration.from_pretrained('Salesforce/blip-image-captioning-large')
```

```python
# descripcion del objeto

modelo_descriptor
```

**Componentes del BlipForConditionalGeneration**


1. **Vision Model (BlipVisionModel)**

Este componente se encarga del procesamiento y codificación de las imágenes.

+ Embeddings (BlipVisionEmbeddings): patch_embedding, una capa convolucional, Conv2d, que convierte las imágenes en una representación vectorial. En este caso, la capa toma una imagen de 3 canales (RGB) y la transforma en una matriz de 1024 características utilizando un tamaño de kernel de 16x16 y un stride de 16x16.

+ Encoder (BlipEncoder):

    + layers (ModuleList): Una lista de 24 capas de BlipEncoderLayer.
    + BlipEncoderLayer: Cada capa contiene:
        + self_attn (BlipAttention): Atención propia, que incluye:
            + dropout: Mecanismo de regularización.
            + qkv: Capa lineal que genera las consultas (queries), claves (keys) y valores (values) para el mecanismo de atención.
            + projection: Proyecta el resultado de la atención en el espacio de características original.
        + layer_norm1 (LayerNorm): Normalización de capas.
        + mlp (BlipMLP): Perceptrón multicapa, que incluye:
            + activation_fn (GELUActivation): Función de activación GELU.
            + fc1 y fc2: Capas lineales para transformar las características.
        + layer_norm2 (LayerNorm): Otra capa de normalización.


+ post_layernorm (LayerNorm): Normalización de capas después del encoder.


2. **Text Decoder (BlipTextLMHeadModel)**

Este componente se encarga de la generación de texto a partir de las características visuales procesadas con un modelo bert (BlipTextModel).

+ Embeddings (BlipTextEmbeddings): Encargado de convertir las secuencias de texto en vectores de características.

    + word_embeddings: Embeddings de palabras con un tamaño de 768.
    + position_embeddings: Embeddings de posiciones para capturar la posición de las palabras en la secuencia.
    + LayerNorm: Normalización de capas.
    + dropout: Mecanismo de regularización.


+ Encoder (BlipTextEncoder):

    + layer (ModuleList): Una lista de 12 capas de BlipTextLayer.
    + BlipTextLayer: Cada capa contiene:
        + attention (BlipTextAttention): Mecanismo de atención.
        + self (BlipTextSelfAttention): Atención propia para el texto.
        + query, key, value: Capas lineales para generar las consultas, claves y valores.
        + dropout: Mecanismo de regularización.
        + output (BlipTextSelfOutput): Proyecta el resultado de la atención y aplica normalización y dropout.
        + crossattention (BlipTextAttention): Atención cruzada para integrar las características visuales.
        + query, key, value: Similar a la atención propia, pero integra características de las imágenes.
        + output (BlipTextSelfOutput): Proyecta el resultado y aplica normalización y dropout.
        + intermediate (BlipTextIntermediate): Intermedia transformaciones de las características.
        + dense: Capa lineal.
        + intermediate_act_fn (GELUActivation): Función de activación GELU.
        + output (BlipTextOutput): Salida del transformador.
        + dense: Capa lineal.
        + LayerNorm: Normalización de capas.
        + dropout: Mecanismo de regularización.


+ **cls (BlipTextOnlyMLMHead)**:

    + predictions (BlipTextLMPredictionHead): Cabeza de predicción del modelo de lenguaje.
    + transform (BlipTextPredictionHeadTransform): Transformaciones previas a la predicción.
    + dense: Capa lineal.
    + transform_act_fn (GELUActivation): Función de activación GELU.
    + LayerNorm: Normalización de capas.
    + decoder: Capa lineal final que convierte las características en predicciones de tokens con un tamaño de 30524.


**Funcionamiento del Modelo**

1. Procesamiento de imágenes: Las imágenes se convierten en vectores de características mediante el Vision Model, utilizando convoluciones y atención propia.


2. Generación de texto: Las características visuales se combinan con embeddings de texto en el Text Decoder para generar descripciones textuales. La atención cruzada permite al modelo integrar información visual directamente en el proceso de generación de texto, mejorando la coherencia y relevancia de las descripciones generadas.


3. Predicción de texto: La cabeza de predicción (BlipTextOnlyMLMHead) transforma las características generadas en tokens de texto, produciendo la descripción final.


```python
# resultado del modelo Blip al darle el vector

tensor = modelo_descriptor.generate(**img_txt_procesado)
```

Le pasamos ahora el tensor generado por el modelo al procesador para que nos devuelva el texto de respuesta.

```python
# respuesta del modelo

procesador.decode(tensor[0], skip_special_tokens=True)
```

Podemos hacerlo de manera no condicional, es decir, no darle ningún texto de entrada al procesador:

```python
# transformación de la imagen 
img_txt_procesado = procesador(images=imagen, return_tensors='pt')


# resultado del modelo Blip al darle el vector
tensor = modelo_descriptor.generate(**img_txt_procesado)


# respuesta del modelo
procesador.decode(tensor[0], skip_special_tokens=True)
```

## 4 - Otro modelo de imagen a texto

Vamos a usar el modelo [TrOCR](https://huggingface.co/microsoft/trocr-base-handwritten) afinado en el [conjunto de datos IAM](https://fki.tic.heia-fr.ch/databases/iam-handwriting-database). Fue introducido en el artículo [TrOCR: Transformer-based Optical Character Recognition with Pre-trained Models](https://arxiv.org/abs/2109.10282) por Li et al. y publicado por Microsoft para extraer textos escritos a mano.

El modelo TrOCR es un modelo de codificador-decodificador, que consiste en un Transformer de imágenes como codificador y un Transformer de texto como decodificador. El codificador de imágenes se inicializó a partir de los pesos de BEiT, mientras que el decodificador de texto se inicializó a partir de los pesos de RoBERTa. Las imágenes se presentan al modelo como una secuencia de parches de tamaño fijo (resolución 16x16), que se incrustan linealmente. Además, se agregan embeddings de posición absoluta antes de alimentar la secuencia a las capas del codificador Transformer. Luego, el decodificador de texto del Transformer genera tokens de manera autorregresiva.


```python
# definimos la tarea y el modelo

tarea = 'image-to-text'  

modelo = 'microsoft/trocr-base-handwritten'
```

```python
# iniciamos el descriptor de imagenes, los modelos se descargan en local, en este caso son unos 1.5Gb

img_pipe = pipeline(task=tarea, model=modelo, max_new_tokens=100)
```

```python
# url de la image

url = 'https://fki.tic.heia-fr.ch/static/img/a01-122-02.jpg'
```

```python
# visualización de la imagen

imagen = Image.open(req.get(url, stream=True).raw)

imagen.resize((1000, 200))
```

```python
# llamada al pipeline para la descripción de la imagen

img_pipe(imagen)
```
