## 1 - Modelos de detección de objetos Zero-Shot

Los modelos de detección de objetos zero-shot son aquellos que pueden identificar y localizar objetos en imágenes, incluso cuando no han sido entrenados específicamente para detectar esos objetos. Utilizan técnicas avanzadas de aprendizaje profundo y se apoyan en información textual para lograr esta capacidad. Aquí se presentan algunos enfoques y modelos destacados en la detección de objetos zero-shot.


**Enfoques en la Detección de Objetos Zero-Shot**

+ Aprendizaje contrastivo: Similar al método utilizado por CLIP, donde se entrenan modelos para aprender representaciones comunes para texto e imágenes en un espacio de embeddings compartido.

+ Transferencia de conocimientos: Utilizar conocimientos previos de modelos preentrenados en grandes conjuntos de datos para transferir estos conocimientos a nuevas tareas y categorías no vistas durante el entrenamiento.

+ Embeddings semánticos: Representar objetos mediante embeddings semánticos (vectores que capturan significados de palabras) y utilizar estos embeddings para relacionar descripciones textuales con regiones de imágenes.


**Modelos Destacados**

1. CLIP (Contrastive Language-Image Pre-Training): Desarrollado por OpenAI, es un modelo que puede ser utilizado para la detección de objetos zero-shot. Aunque CLIP no está diseñado específicamente para detección de objetos, su capacidad para asociar texto e imágenes puede ser aprovechada en esta tarea. CLIP se entrena para asociar descripciones textuales con imágenes a través de un aprendizaje contrastivo. Para la detección de objetos, CLIP puede ser combinado con métodos de segmentación de imágenes para localizar regiones que correspondan a una descripción textual específica.


2. ViLD (Vision-Language Detr): Es un modelo desarrollado por Google Research que extiende el modelo DETR (Detection Transformer) para tareas de detección de objetos zero-shot. ViLD utiliza un modelo preentrenado similar a CLIP para generar embeddings de texto e imágenes y luego entrena un modelo DETR para detectar objetos en imágenes, incluso aquellos no vistos durante el entrenamiento.


3. ZSD (Zero-Shot Detection): Varios modelos específicos de detección zero-shot han sido propuestos en la literatura, basándose en técnicas de aprendizaje profundo y embeddings semánticos. Estos modelos generalmente se entrenan en un conjunto de datos con anotaciones de categorías conocidas y utilizan información textual para generalizar a nuevas categorías.


**Desafíos**

+ Generalización: Uno de los principales desafíos es asegurar que los modelos puedan generalizar efectivamente a categorías no vistas sin pérdida significativa de precisión.

+ Escalabilidad: A medida que se aumentan las categorías de objetos, mantener la precisión y eficiencia del modelo es crucial.

+ Interacción multimodal: Mejorar la interacción entre texto e imágenes para una detección más precisa y robusta.


## 2 - Pipeline de Transformers para detección de objetos Zero-Shot

Vamos a usar el modelo [OwlViT](https://huggingface.co/google/owlvit-base-patch32) (Open-World Localization Vision Transformer) que fue propuesto en [Simple Open-Vocabulary Object Detection with Vision Transformers](https://arxiv.org/abs/2205.06230) por Matthias Minderer et al. OwlViT es un modelo de detección de objetos condicionado por texto en Zero-Shot que puede utilizarse para consultar una imagen con una o múltiples consultas de texto. El modelo utiliza CLIP como su columna vertebral multimodal, con un Transformer similar a ViT para obtener características visuales y un modelo de lenguaje causal para obtener las características del texto. Para usar CLIP para la detección, OwlViT elimina la capa de pooling de tokens finales del modelo de visión y añade una cabeza ligera de clasificación y de caja a cada token de salida del Transformer. La clasificación de vocabulario abierto se habilita reemplazando los pesos de la capa de clasificación fija con los embeddings de nombres de clase obtenidos del modelo de texto. Los autores primero entrenan CLIP desde cero y lo ajustan de extremo a extremo con las cabezas de clasificación y de caja en conjuntos de datos de detección estándar utilizando una pérdida de emparejamiento bipartito. Se pueden usar una o múltiples consultas de texto por imagen para realizar la detección de objetos condicionada por texto en Zero-Shot.


### 2.1 - Uso del pipeline

```python
# logging para quitar warnings de actualización de pesos del modelo

from transformers import logging

logging.set_verbosity_error()
```

```python
# importamos desde la librería transformers el pipeline

from transformers import pipeline
```

```python
# definimos la tarea y el modelo

tarea = 'zero-shot-object-detection'  

modelo = 'google/owlvit-base-patch32'
```

```python
# iniciamos el clasificador de imagenes, los modelos se descargan en local, en este caso son unos 615Mb

img_pipe = pipeline(task=tarea, model=modelo)
```

Vamos a usar una imagen de [4 cachorros](https://i0.wp.com/planetamascotaperu.com/wp-content/uploads/2023/09/cachorros1.jpg) subidos en una verja para usar este modelo.

```python
# imagen de cachorros

url = 'https://i0.wp.com/planetamascotaperu.com/wp-content/uploads/2023/09/cachorros1.jpg'
```

```python
# visualizamos la imagen con PIL en 600x400 píxeles

from PIL import Image
import requests as req

imagen = Image.open(req.get(url, stream=True).raw)
```

```python
# visualización 

imagen
```

```python
# etiquetas candidatas, ¿qué hay en la imagen?

etiquetas = ['parrot', 'cat', 'dog']
```

```python
# llamada al pipeline para la clasificación de la imagen zero-shot

respuesta = img_pipe(image=imagen, candidate_labels=etiquetas)

respuesta
```

La respuesta de pipeline es igual que un modelo de detección de objetos, una lista de diccionarios con claves: probabilidad, etiqueta y caja, donde cada caja es a su vez un diccionario que contiene las coordenadas del objeto detectado.

### 2.2 - Pintando las cajas en la imagen

Una vez que hemos obtenido las cajas que contienen las coordenadas de los objetos detectados, podemos pintar rectángulos en la imagen para y dejar los objetos detectados definados en la misma. También pondremos la etiqueta escrita sobre la caja.

```python
# importamos los objetos para pintar

from PIL import ImageDraw, ImageFont
```

### 3.1 - Procesador de imágenes

AutoProcessor es una clase de la biblioteca transformers diseñada para proporcionar una interfaz unificada y conveniente para preprocesar diferentes tipos de datos, como texto, imágenes y audio, antes de ingresarlos en un modelo. Es parte de la familia de clases "Auto" de Hugging Face, que permiten la carga automática de modelos y procesadores correspondientes, reduciendo la complejidad y el esfuerzo manual requerido para configurar estos componentes.


**Características de AutoProcessor**

1. Interfaz unificada: Proporciona una interfaz común para preprocesar diversos tipos de datos, lo que simplifica el uso de modelos multimodales.


2. Detección automática: Detecta automáticamente el tipo de procesador necesario en función del modelo cargado, evitando la necesidad de especificar manualmente el procesador correspondiente.


3. Compatibilidad multimodal: Soporta texto, imágenes, y audio, permitiendo el procesamiento de entradas multimodales.

```python
# transforma las imágenes 

procesador = AutoProcessor.from_pretrained('google/owlvit-base-patch32')
```

```python
# descripcion del objeto

procesador
```

**Componentes del OwlViTProcessor**


1. **Image Processor** (OwlViTImageProcessor):

    + crop_size: Tamaño de recorte de la imagen, especificado como un diccionario con keys ancho y alto, ambos de 768 píxeles.
    
    + do_center_crop: Indica si se debe hacer un recorte centrado. Aquí está configurado como false.
    
    + do_convert_rgb: Convierte la imagen a formato RGB, configurado como true.
    
    + do_normalize: Normaliza la imagen, configurado como true.
    
    + do_rescale: Reescala la imagen, configurado como true.
    
    + do_resize: Redimensiona la imagen, configurado como true.
    
    + image_mean: Valores de media para normalización, especificados como una lista de tres valores.
    
    + image_processor_type: Tipo del procesador de imágenes.
    
    + image_std: Valores de desviación estándar para normalización, especificados como una lista de tres valores.
    
    + resample: Método de remuestreo, aquí es 3 que corresponde a Image.BICUBIC en PIL.
    
    + rescale_factor: Factor de reescalado, aquí es 0.00392156862745098 (1/255).
    
    + size: Tamaño de la imagen redimensionada, especificado como un diccionario con keys ancho y alto, ambos de 768 píxeles.
    


2. **Tokenizer** (CLIPTokenizerFast):

    + name_or_path: Nombre o ruta del tokenizador.
    
    + vocab_size: Tamaño del vocabulario, aquí es 49408.
    
    + model_max_length: Longitud máxima del modelo, aquí es 16.
    
    + is_fast: Indica si es un tokenizador rápido, configurado como true.
    
    + padding_side: Lado de relleno, aquí es 'right'.
    
    + truncation_side: Lado de truncamiento, aquí es 'right'.
    
    + special_tokens: Diccionario con tokens especiales como bos_token, eos_token, unk_token, y pad_token.
    
    + clean_up_tokenization_spaces: Limpieza de espacios de tokenización, configurado como true.
    
    + added_tokens_decoder: Diccionario con tokens añadidos decodificados, que incluye:
        + 0: Token "!".
        + 49406: Token "<|startoftext|>" (inicio de secuencia).
        + 49407: Token "<|endoftext|>" (fin de secuencia).




OwlViTProcessor combina el procesamiento de imágenes y el tokenizador para manejar entradas multimodales, imágenes y texto. Este procesador asegura que las imágenes y el texto se preparen de manera adecuada antes de ser ingresados en el modelo OwlViT.

### 3.2 - Modelo detector de objetos Zero-Shot

AutoModelForZeroShotObjectDetection es una clase de la biblioteca transformers diseñada para facilitar la carga y el uso de modelos preentrenados para la detección de objetos en zero-shot. Proporciona una interfaz sencilla para cargar diferentes modelos de detección de objetos en zero-shot de la biblioteca transformers.


```python
# inicializacion del modelo OwlViT

modelo_detector = AutoModelForZeroShotObjectDetection.from_pretrained('google/owlvit-base-patch32')
```

```python
# descripcion del objeto

modelo_detector
```

**Descripción de la arquitectura**:


1. **OwlViTModel**:

+ **Text Model**:

    + Embeddings:
        + token_embedding: Capa de embedding para tokens de texto, con una dimensión de 49408 para el vocabulario y 512 para los embeddings.
        + position_embedding: Capa de embedding para las posiciones de los tokens, con una dimensión de 16 para la longitud máxima de la secuencia y 512 para los embeddings.

    + Encoder: Compuesto por 12 capas (0-11) de OwlViTEncoderLayer, cada una con:
        + self_attn (Self-Attention): Mecanismo de atención con proyecciones k_proj, v_proj, q_proj y out_proj, capas de proyección lineal, todas con dimensiones de entrada y salida de 512.
        + layer_norm1: Normalización por capas con dimensión de 512.
        + mlp (Multilayer Perceptron): Contiene una función de activación QuickGELUActivation, una variante optimizada de la función de activación GELU, y dos capas totalmente conectadas, fc1 y fc2, con dimensiones de 512 a 2048 y 2048 a 512, respectivamente.
        + layer_norm2: Otra normalización por capas con dimensión de 512.

    + final_layer_norm: Normalización por capas final con dimensión de 512.


+ **Vision Model**:

    + Embeddings:
        + patch_embedding: Capa de convolución, Conv2d, para crear embeddings de paquetes de imágenes, con 3 canales de entrada y 768 de salida, y un tamaño de kernel de 32x32 con stride de 32x32.
        + position_embedding: Capa de embedding para las posiciones de los paquetes en la imagen, con una dimensión de 577 y 768 para los embeddings.

    + Pre-LayerNorm: Normalización por capas previa con dimensión de 768.
    
    + Encoder: Compuesto por 12 capas (0-11) de OwlViTEncoderLayer, cada una con:
        + self_attn (Self-Attention): Mecanismo de atención con proyecciones k_proj, v_proj, q_proj y out_proj, todas con dimensiones de entrada y salida de 768.
        + layer_norm1: Normalización por capas con dimensión de 768.
        + mlp (Multilayer Perceptron): Contiene una función de activación QuickGELUActivation, y dos capas totalmente conectadas, fc1 y fc2, con dimensiones de 768 a 3072 y 3072 a 768, respectivamente.
        + layer_norm2: Otra normalización por capas con dimensión de 768.
    
    + post_layernorm: Normalización por capas posterior con dimensión de 768.
    + Proyecciones:
        + visual_projection: Proyección lineal de características visuales, de 768 a 512.
        + text_projection: Proyección lineal de características textuales, de 512 a 512.


2. **Cabeza de clasificación**:

    + Dense Layers: dense0, capa densa de 768 a 512.
    + Logit Layers:
        + logit_shift: Capa lineal de 768 a 1.
        + logit_scale: Capa lineal de 768 a 1.
    + Activación: Función de activación ELU (Exponential Linear Unit) con alfa, valor que controla la amplitud de la parte negativa de la función, de 1.0.


3. **Cabeza de predicción de cajas**:

    + Dense Layers:
        + dense0: Capa densa de 768 a 768.
        + dense1: Capa densa de 768 a 768.
    + Activación: Función de activación GELU (Gaussian Error Linear Unit).
    + Output Layer: dense2, capa lineal de 768 a 4, para las coordenadas de las cajas delimitadoras.


4. **Normalización y activación**:

    + Layer Norm: Normalización por capas con dimensión de 768.
    + Sigmoid: Función de activación sigmoide para producir salidas en el rango de 0 a 1.


**Funcionamiento General**

1. Entrada de Texto e Imagen: El modelo acepta una imagen y descripciones textuales de los objetos a detectar.

2. Extracción de Características: Las características textuales se extraen utilizando el text_model y las características visuales se extraen utilizando el vision_model.

3. Proyección de Características: Las características visuales y textuales se proyectan a un espacio común.

4. Clasificación y Localización:
    + La cabeza de clasificación (class_head) se encarga de predecir la presencia de los objetos descritos.
    + La cabeza de predicción de cajas (box_head) se encarga de predecir las coordenadas de las cajas delimitadoras para los objetos detectados.

5. Normalización y Activación: Se aplican normalización y funciones de activación para ajustar las predicciones finales.

```python
# resultado del modelo OwlViT al darle el vector

tensor = modelo_detector(**img_procesada)
```

Con el tensor de salida del modelo, se lo devolvemos al procesador para que nos devuelva las etiquetas, sus probabilidades y las cajas que contienen los objetos. Para ello le vamos a dar al procesador el tensor de salida del modelo, el tamaño de salida de la imagen y un umbral de probabilidad para que el objeto sea detectado, en este caso lo pondremos en 0.1 puesto que las probabilidades son bajas.

```python
# resultado post-procesado

resultado = procesador.post_process_object_detection(tensor, 
                                                     threshold=0.1, 
                                                     target_sizes=[imagen.size[::-1]])[0]

resultado
```

Vemos que en el resultado tenemos las cajas, las probabilidades y las etiquetas. Las etiquetas vienen en formato numérico, son los índices de la lista de etiquetas que hemos definido. Vamos a hacer un bucle sobre el resultado para ver las etiquetas, probabilidades y los píxeles de las cajas.

```python
# hacemos un bucle sobre el resultado

for prob, eti, caja in zip(resultado['scores'], resultado['labels'], resultado['boxes']):
    
    # redondeamos y convertimos a lista los datos de los píxeles de la caja
    caja = [round(i, 2) for i in caja.tolist()]
    
    # creamos un print llamando al diccionario de etiquetas 
    print(f'Detectado {etiquetas[eti.item()]} con probabilidad {round(prob.item(), 3)} en {caja}')
```

### 3.3 - Resumen del código

Vamos a poner todo el código junto en una sola función:

```python
# librerías
from transformers import AutoProcessor, AutoModelForZeroShotObjectDetection



def detector_zero_shot(imagen: object, etiquetas: list) -> list:
    
    """
    Función para detectar objetos en imagenes con modelo OwlViT Zero-Shot
    
    Params:
    imagen: objeto PIL.JpegImagePlugin.JpegImageFile
    etiquetas: lista de etiquetas definidas por el usuario
    
    Return:
    lista de strings que contienen la etiqueta, probabilidad y posiciones de las cajas
    """
    
    # transforma las imágenes 
    procesador = AutoProcessor.from_pretrained('google/owlvit-base-patch32')
    
    
    # transformación de la imagen dándole las etiquetas que hemos definido
    img_procesada = procesador(text=etiquetas, images=imagen, return_tensors='pt')
    
    
    # inicializacion del modelo OwlViT
    modelo_detector = AutoModelForZeroShotObjectDetection.from_pretrained('google/owlvit-base-patch32')
    
    
    # resultado del modelo OwlViT al darle el vector
    tensor = modelo_detector(**img_procesada)
    
    
    # resultado post-procesado
    resultado = procesador.post_process_object_detection(tensor, 
                                                         threshold=0.1, 
                                                         target_sizes=[imagen.size[::-1]])[0]
    
    
    # lista de salida de la función
    salida = []
    
    
    # hacemos un bucle sobre el resultado
    for prob, eti, caja in zip(resultado['scores'], resultado['labels'], resultado['boxes']):

        # redondeamos y convertimos a lista los datos de los píxeles de la caja
        caja = [round(i, 2) for i in caja.tolist()]

        # creamos un print llamanda al diccionario de etiquetas
        salida.append({'etiqueta': etiquetas[eti.item()],
                       'probabilidad': round(prob.item(), 3),
                       'caja': caja})
        
     
    return salida
```
